```python
## Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##Loading the dataset
df = pd.read_csv("customer_shopping_behavior.csv")
```

```python
## Checking the first few rows of the dataset
df.head()

```

```python
## Checking for missing values
df.isna().sum()
```

```python
## Data Analysis
df.describe()
```

```python
## Handling missing values in "Review Rating" by filling them with the median rating for each category
df["Review Rating"] = df.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median()))
```

```python
## Checking the first few rows of the dataset after handling missing values
df.head(10)
df.isna().sum()
```

```python
## Categorizing "Age" into age groups
labels = ["Young Adult", "Adult", "Middle Age", "Senior"]
df["Age Group"] = pd.qcut(df["Age"], q=4, labels=labels)
```

```python
## Checking the first few rows of the dataset after categorizing age groups
df.head()   
```

```python
## Changing column names to lowercase and replacing spaces with underscores for better readability
df.columns = df.columns.str.lower().str.replace(" ", "_")   
```

```python
## Checking the updated column names
df.columns
```

```python

## Renaming "purchase_amount_(usd)" to "purchase_amount" for better readability
df.rename(columns={"purchase_amount_(usd)":"purchase_amount"}, inplace=True)
```

```python
## Checking the updated column names after renaming
df.columns
```

```python
## Exploring the relationship between age and purchase amount
df[["age", "age_group"]].head(10)
```

```python
## Visualizing the distribution of purchase amounts across different age groups
df.head(5)
```

```python
## creating a mapping of frequency to number of days for better analysis
frequency_mapping = {
    "Fortnightly": 14,
    "Weekly": 7,
    "Biweekly": 14,
    "Monthly": 30,
    "Quarterly": 90,
    "Annually": 365,
    "Every 2 months": 60,
    "Every 3 months": 90,
    "Every 6 months": 180,
    "Every 12 months": 365
}
```

```python
## Mapping the frequency of purchases to the number of days between purchases
df["purchase_frequency_days"] = df["frequency_of_purchases"].map(frequency_mapping)
```

```python
## Checking the first few rows of the dataset after mapping purchase frequency to days
df[["purchase_frequency_days", "frequency_of_purchases"]].head(10)
```

```python
## Checking the data types of the columns to ensure they are appropriate for analysis
df.info()
```

```python
## Converting "purchase_frequency_days" to integer type for better analysis
df["purchase_frequency_days"] = df["purchase_frequency_days"].astype("Int64")
```

```python
## Checking the data types of the columns to ensure they are appropriate for analysis
df.info()
```

```python
## Checking the first few rows of the dataset after changing purchase_frequency_days data type to integer
df[["purchase_frequency_days", "frequency_of_purchases"]].head(10)
```

```python
## Visualizing the relationship between purchase frequency and purchase amount
df.head(10)
```

```python
## Exploring the relationship between discount applied and promo code used
df[["discount_applied", "promo_code_used"]].head(10)
```

```python
## Checking if there is a correlation between discount applied and promo code used
(df["discount_applied"] == df["promo_code_used"]).all()
```

```python
## Since there is a correlation between discount applied and promo code used, we can drop the "discount_applied" column to avoid redundancy in our analysis 
df.drop("discount_applied", axis=1, inplace=True)
```

```python
##checking if the column is dropped
df.head(10)
```

```python
## Exporting the cleaned dataset to a SQL Server database
from sqlalchemy import create_engine
from urllib.parse import quote_plus

host = "localhost"
database = "customer_behaviour_project"

driver = quote_plus("ODBC Driver 17 for SQL Server")

engine = create_engine(
    f"mssql+pyodbc://@{host}/{database}?driver={driver}&trusted_connection=yes"
)

table_name = "customer_behaviour"

df.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"Data has been successfully inserted into the '{table_name}' table in the '{database}' database.")
```
